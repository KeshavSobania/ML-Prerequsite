Question 1: 
* Total No. of Multiplications are  (nmk)
* Total No. of Additions are  (nm(k-1))
* Explaintation : In this scenario, determining each element of the resulting matrix involves performing \( k \) multiplications, leading to a total of ( nmk ) multiplications overall. Additionally, since \( k-1 \) additions are needed for each element, the total number of additions required is ( nm(k-1) ).

```python
# Question 2

import time

def multiply_matrices(A, B):
    n = len(A)
    m = len(A[0])
    p = len(B[0])
    result = [[0] * p for i in range(n)]
    
    for i in range(n):
        for j in range(p):
            for k in range(m):
                result[i][j] += A[i][k] * B[k][j]
                
    return result

A = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
B = [[9, 8, 7], [6, 5, 4], [3, 2, 1]]

start_time = time.time()
result = multiply_matrices(A, B)
end_time = time.time()
time = end_time - start_time

print("\nResult using list of lists:")
print(result)
print(f"Time taken using list of lists: {time} seconds")
```

NumPy is faster than the list of list method because the list of list method multiplies elements one by one, which takes a lot of time and numPy processes whole arrays at once to do it faster.


```python
# Queestion 3
import time
n=len(arr)
def median(arr) :
    for i in range(len(arr)//2 -1) :
        max = arr[0] 
        for i in range(len(arr)) :
            if(arr[i] > max) :
                max = arr[i]
        arr.remove(max)

    m1 = arr[0]
    for i in range(len(arr)):
        if(arr[i] > m1) :
            m1 = arr[i]

    arr.remove(m1)
    m2 = arr[0]
    for i in range(len(arr)):
        if(arr[i] > m2) :
            m2 = arr[i]
    median = (m1 + m2)/2
    
    if(n%2 != 0) :
        median = m2 
    else :
        median = (m1+m2)/2
    
    return median

arr = list(range(1,100))

start_time = time.time()
median = median(arr)
end_time = time.time() 
time=end_time-start_time

print(f"The Median is {median} ")    
print(f"Time taken this method is {time}")

#O(n^2) is time complexity
```

```python
# alternate method by sorting the array
import time
    
def median(arr):
    arr.sort()
    n = len(arr)
    if n % 2 != 0:
        median = arr[n//2]
    else:
        median = (arr[n//2 - 1] + arr[n//2]) / 2
        
    return median

arr = list(range(1,100))
start_time = time.time()
median = median(arr)
end_time = time.time()
time=end_time-start_time

print(f"By using sorting method Median is {median}.")  
print(f"Time taken by this method is {time}")

#O(nlogn) is time complexity
```
